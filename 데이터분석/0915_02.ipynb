```python
#클래스 (틀)
class abcdefg:
  aa=1
  def __init__(self,a={0,0},b=0,c=-1.5,d=[0],e=(0),f={0},g=bool(0)): 
    self.a=a
    self.b=b
    self.c=c
    self.d=d
    self.e=e
    self.f=f
    self.g=g
  def printtype(self):
    for z in [(x[0],x[1],type(x[1])) for x in list(self.__dict__.items())]:
      print(z)
  def printglobalvariable(self):
    global h
    h+=1 #h=h+1
    print(h,end=' ')
print(abcdefg().a)
print(abcdefg.aa)
abcdefg(a=a,b=b,c=c,d=d,e=e,f=f,g=g).printtype()
```

```python
#global 변수, range, 함수
print('1st: ',end='')
for x in range(len(c+c)):  
  abcdefg().printglobalvariable()
print()

def justfunc():
  print('2nd: ',end='')
  for x in range(len(c+c)):
    print('%2d' %b, end=' ')
  print()
justfunc()

print('3rd:',h)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=Image.open('/g6.jpg').convert('L') #코드 돌려볼 시, 원하는 사진 업로드 후 이름 바꿀 것
imagetwo=np.asarray(image.resize((256,256)))
imagefour=[]
for x in imagetwo: #사칙연산으론 Normalized된 결과를 출력하는 imshow로 인해 쉽게 바뀌지 않음.
  arr=[]
  for y in x:
    arr.append(y**5)
  imagefour.append(arr)
plt.imshow(imagefour,cmap='gray')
```
